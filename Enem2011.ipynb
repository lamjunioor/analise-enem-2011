```
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('http://aulas.artificial.eng.br/Home/pgca008-mineracao-de-dados/material/ENEM_MINERACAO_v2.csv')
fig = plt.figure(figsize=(12, 12)) #para melhorar a visualização da dispersão aumentei o valor para 90,90 porém coloquei 12, 12 novamente para a visualização inicial
ax  = fig.add_subplot(1, 1, 1)

ax.scatter(data.NU_NT_MT, data.NU_NT_CH)

ax.set_xlabel("Matemática")
ax.set_ylabel("Ciências Humanas")

ax.set_title("Gráfico de dispersão das notas das provas de matemática x ciencias humanas")

plt.show()
```

O gráfico de dispersão gerado nos mostra uma relação positiva fraca entre as notas de matemática e ciências humanas nas notas do ENEM 2011 o que significa que apesar de possuir uma relação entre o aumento das notas ele não é tão acentuado. 

Porém uma observação importante (que é melhor observada aumentando o valor em figsize) é que apesar de haver um número baixo de pessoas que tiraram nota alta somente em Ciências Humanas esse número é ainda menor comparado às pessoas que tiraram nota alta somente em Matemática. O que leva a acreditar que há uma maior dificuldade em matemática.

```
fig = plt.figure(figsize=(12, 9))
ax  = fig.add_subplot(1, 1, 1)

data_sp= data.loc[data['UF_INSC'] == 'SP']
data_mg= data.loc[data['UF_INSC'] == 'MG']
data_ba= data.loc[data['UF_INSC'] == 'BA']

kwargs = dict(histtype='stepfilled', alpha=0.7)

ax.hist(data_sp.NU_NT_MT, **kwargs)
ax.hist(data_mg.NU_NT_MT, **kwargs)
ax.hist(data_ba.NU_NT_MT, **kwargs)


plt.show()
```

Através do gráfico é possível visualizar que os estados de SP e MG possuem gráficos semelhantes, a diferença é dada através da quantidade de alunos que fizeram o ENEM nesses estados. SP, por ter uma população maior é mostrado com um número grande de notas, MG tem uma quantidade menor de alunos porém a porcentagem das notas está bem semelhante à de SP.

O estado da Bahia tem uma quantidade menor de alunos que fizeram o ENEM 2011, porém, mantendo a proporcionalidade entre estados vemos um baixo desempenho dos alunos nesse estado. Diferente de SP e MG, onde a nota mais comum foi 600, a BA teve seu pico na nota 400.
